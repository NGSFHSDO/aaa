# @ 표시 API 전체 수집

## 공통 수집 함수

```python
import json
import math
import os
import sqlite3
import time
from pathlib import Path

import pandas as pd
import requests
from dotenv import load_dotenv

load_dotenv()
SERVICE_KEY = os.getenv("SERVICE_KEY_DECODED")
DATA_DIR = Path("data/kotra")
DATA_DIR.mkdir(parents=True, exist_ok=True)


def parse_total_and_items(data):
    if "response" in data:
        body = data["response"]["body"]
        total_count = int(body["totalCnt"])
        items = body.get("itemList", {}).get("item", [])
    else:
        total_count = int(data["totalCount"])
        items = data.get("records", [])

    if isinstance(items, dict):
        items = [items]
    return total_count, items


def serialize_value(value):
    if isinstance(value, (dict, list)):
        return json.dumps(value, ensure_ascii=False)
    return value


def collect_api_data(url, params, label, rename_cols, table_name, output_stem):
    first_res = requests.get(url, params=params, timeout=30)
    first_res.raise_for_status()
    first_data = first_res.json()

    total_count, _ = parse_total_and_items(first_data)
    num_rows = int(params["numOfRows"])
    total_pages = math.ceil(total_count / num_rows)

    print(f"[{label}] 전체 건수:", total_count)
    print(f"[{label}] 전체 페이지:", total_pages)

    all_items = []

    for page in range(1, total_pages + 1):
        page_params = params.copy()
        page_params["pageNo"] = page
        success = False

        for attempt in range(3):
            try:
                res = requests.get(url, params=page_params, timeout=30)
                res.raise_for_status()
                data = res.json()
                _, items = parse_total_and_items(data)
                all_items.extend(items)

                print(f"[{label}] {page}/{total_pages} 페이지 완료")
                success = True
                break

            except Exception as e:
                print(f"[{label}] {page}페이지 오류, {attempt + 1}번째 재시도:", e)
                time.sleep(2)

        if not success:
            print(f"[{label}] {page}페이지는 최종 실패. 다음 페이지로 넘어갑니다.")

        time.sleep(0.3)

    df = pd.DataFrame(all_items)
    df = df.rename(columns=rename_cols)
    df = df.apply(lambda col: col.map(serialize_value))

    db_path = DATA_DIR / f"{output_stem}.db"
    csv_path = DATA_DIR / f"{output_stem}.csv"

    conn = sqlite3.connect(db_path)
    df.to_sql(table_name, conn, if_exists="replace", index=False)
    conn.close()

    df.to_csv(csv_path, index=False, encoding="utf-8-sig")

    print(f"[{label}] 저장 완료")
    print(f"[{label}] DB:", db_path)
    print(f"[{label}] CSV:", csv_path)
    print(f"[{label}] 수집 행 수:", len(df))
    return df
```

```python
# 첨부파일목록에서 PDF 파일명과 URL을 추출한 뒤 다운로드
import json
import re
import sqlite3
import time
from pathlib import Path

import pandas as pd
import requests
from tqdm.auto import tqdm

DB_PATH = Path("data/kotra/globalSupplyInsights.db")
TABLE_NAME = "globalSupplyInsights"
PDF_DIR = Path("data/kotra/globalSupplyInsights_pdfs")
PDF_DIR.mkdir(parents=True, exist_ok=True)


def safe_filename(name):
    name = str(name).strip() or "download.pdf"
    name = re.sub(r"[\\/:*?\"<>|]", "_", name)
    name = re.sub(r"\s+", " ", name).strip()
    if not name.lower().endswith(".pdf"):
        name += ".pdf"
    return name


def extract_pdf_files(row):
    raw = row.get("첨부파일정보목록", "")
    if not raw:
        return []

    try:
        data = json.loads(raw)
    except json.JSONDecodeError:
        return []

    files = data.get("realAtfileInfo", [])
    if isinstance(files, dict):
        files = [files]

    pdf_files = []
    for file_info in files:
        file_name = file_info.get("realAtfileName", "")
        file_url = file_info.get("realAtfileUrl", "")
        if file_url and file_name.lower().endswith(".pdf"):
            pdf_files.append({
                "게시글일련번호": row.get("게시글일련번호", ""),
                "게시글제목": row.get("게시글제목", ""),
                "공개일시": row.get("공개일시", ""),
                "파일명": file_name,
                "파일URL": file_url,
            })
    return pdf_files


conn = sqlite3.connect(DB_PATH)
df = pd.read_sql(f"SELECT * FROM {TABLE_NAME}", conn)
conn.close()

pdf_rows = []
for _, row in df.iterrows():
    pdf_rows.extend(extract_pdf_files(row))

pdf_df = pd.DataFrame(pdf_rows)
print("PDF 파일 수:", len(pdf_df))

download_logs = []

for _, row in tqdm(pdf_df.iterrows(), total=len(pdf_df), desc="PDF 다운로드"):
    ntt_sn = str(row.get("게시글일련번호", "")).strip()
    file_name = safe_filename(row.get("파일명", ""))
    save_name = f"{ntt_sn}_{file_name}" if ntt_sn else file_name
    save_path = PDF_DIR / save_name

    if save_path.exists() and save_path.stat().st_size > 0:
        download_logs.append({**row.to_dict(), "저장경로": str(save_path), "상태": "skip"})
        continue

    success = False
    for attempt in range(3):
        try:
            res = requests.get(row["파일URL"], timeout=60)
            res.raise_for_status()
            save_path.write_bytes(res.content)
            download_logs.append({**row.to_dict(), "저장경로": str(save_path), "상태": "success"})
            success = True
            break
        except Exception as e:
            if attempt == 2:
                download_logs.append({**row.to_dict(), "저장경로": str(save_path), "상태": f"fail: {e}"})
            else:
                time.sleep(2)

    time.sleep(0.2)

download_log_df = pd.DataFrame(download_logs)
download_log_path = PDF_DIR / "download_log.csv"
download_log_df.to_csv(download_log_path, index=False, encoding="utf-8-sig")

print("다운로드 폴더:", PDF_DIR)
print("다운로드 로그:", download_log_path)
download_log_df["상태"].value_counts(dropna=False)
```

# HS 코드 한글 품목명 merge

`data/HS코드(2025).xls`를 기준으로 KOTRA 데이터의 HS 코드에 한글 품목명 후보를 붙입니다.

- `exportRecommendInfo`: `품목코드` 기준
- `importRegItems`: `HSCD` 기준
- `ovseaAuthInfo`: `HSCD` 기준

원본 파일은 그대로 두고, 결과는 `data/kotra/*_hs.db`, `data/kotra/*_hs.csv`로 저장합니다.
추가 컬럼은 `HS정리코드`, `HS매칭레벨`, `HS한글품목명목록` 3개입니다.


```python
import re
import sqlite3
from pathlib import Path

import pandas as pd
from tqdm.auto import tqdm

DATA_DIR = Path("data/kotra")
HS_2025_PATH = next(Path("data").glob("HS*2025*.xls"))
WCO_PATH = Path("data/WCO 2017 2022 Correlation.xlsx")

TARGETS = [
    {"stem": "exportRecommendInfo", "table": "exportRecommendInfo", "hs_col": "품목코드"},
    {"stem": "importRegItems", "table": "importRegItems", "hs_col": "HSCD"},
    {"stem": "ovseaAuthInfo", "table": "ovseaAuthInfo", "hs_col": "HSCD"},
]


def clean_hs_code(value):
    if pd.isna(value):
        return ""
    return re.sub(r"\D", "", str(value)).strip()


def unique_join(values):
    result = []
    seen = set()
    for value in values:
        if pd.isna(value):
            continue
        text = str(value).strip()
        if not text or text in seen:
            continue
        seen.add(text)
        result.append(text)
    return ", ".join(result)


# 2025년 HS 10자리 기준표 로드
hs = pd.read_excel(HS_2025_PATH, dtype=str, keep_default_na=False)
hs.columns = hs.columns.str.strip()
hs["HS10"] = hs["HS부호"].map(clean_hs_code).str.zfill(10)
hs = hs[hs["HS10"].str.len() == 10].copy()
hs["HS6"] = hs["HS10"].str[:6]
hs["HS4"] = hs["HS10"].str[:4]

hs10_name_map = hs.groupby("HS10", sort=False)["한글품목명"].apply(unique_join).to_dict()
hs6_name_map = hs.groupby("HS6", sort=False)["한글품목명"].apply(unique_join).to_dict()
hs4_name_map = hs.groupby("HS4", sort=False)["한글품목명"].apply(unique_join).to_dict()

# WCO 2017 -> 2022 HS6 보정표 로드
wco = pd.read_excel(WCO_PATH, sheet_name="WITH DESCS", dtype=str, keep_default_na=False)
wco["V2017_6"] = wco["V2017"].map(clean_hs_code).str.zfill(6)
wco["V2022_6"] = wco["V2022"].map(clean_hs_code).str.zfill(6)
wco = wco[(wco["V2017_6"].str.len() == 6) & (wco["V2022_6"].str.len() == 6)]
wco_map = wco.groupby("V2017_6", sort=False)["V2022_6"].apply(lambda s: list(dict.fromkeys(s))).to_dict()


def match_hs_korean_names(raw_code):
    code = clean_hs_code(raw_code)

    if len(code) >= 10:
        hs10 = code[:10]
        if hs10 in hs10_name_map:
            return code, "HS10", hs10_name_map[hs10]

    if len(code) >= 6:
        hs6 = code[:6]
        if hs6 in hs6_name_map:
            return code, "HS6", hs6_name_map[hs6]

        wco_candidates = wco_map.get(hs6, [])
        names = unique_join(hs6_name_map[c] for c in wco_candidates if c in hs6_name_map)
        if names:
            return code, "WCO_HS6", names

    if len(code) >= 4:
        hs4 = code[:4]
        if hs4 in hs4_name_map:
            return code, "HS4", hs4_name_map[hs4]

    return code, "unmatched", ""


for target in TARGETS:
    stem = target["stem"]
    table = target["table"]
    hs_col = target["hs_col"]

    input_db = DATA_DIR / f"{stem}.db"
    output_db = DATA_DIR / f"{stem}_hs.db"
    output_csv = DATA_DIR / f"{stem}_hs.csv"
    output_table = f"{stem}_hs"

    conn = sqlite3.connect(input_db)
    df = pd.read_sql(f"SELECT * FROM {table}", conn)
    conn.close()

    matches = [
        match_hs_korean_names(value)
        for value in tqdm(df[hs_col], desc=f"{stem} HS merge")
    ]
    match_df = pd.DataFrame(
        matches,
        columns=["HS정리코드", "HS매칭레벨", "HS한글품목명목록"],
    )

    df_merged = pd.concat([df, match_df], axis=1)

    conn = sqlite3.connect(output_db)
    df_merged.to_sql(output_table, conn, if_exists="replace", index=False)
    conn.close()

    df_merged.to_csv(output_csv, index=False, encoding="utf-8-sig")

    print(f"저장 완료: {output_db} / table={output_table}")
    print(f"저장 완료: {output_csv}")
    print(df_merged[[hs_col, "HS정리코드", "HS매칭레벨", "HS한글품목명목록"]].head(3))
    print()
```

# 기업성공사례 임베딩

```python
import gc
import json
import sqlite3
from pathlib import Path

import faiss
import mlx.core as mx
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from mlx_embeddings import load
from tqdm.auto import tqdm

MODEL_NAME = "majentik/Qwen3-Embedding-0.6B-MLX-4bit"
DB_PATH = Path("data/kotra/compSucsCase.db")
TABLE_NAME = "compSucsCase"
OUT_DIR = Path("data/faiss/compSucsCase_qwen3_embedding")
BATCH_DIR = OUT_DIR / "batches"
INDEX_PATH = OUT_DIR / "index.faiss"
METADATA_PATH = OUT_DIR / "metadata.parquet"
CONFIG_PATH = OUT_DIR / "config.json"

BATCH_SIZE = 1
MAX_LENGTH = 32768
TEXT_COL = "본문내용"

OUT_DIR.mkdir(parents=True, exist_ok=True)
BATCH_DIR.mkdir(parents=True, exist_ok=True)

embedding_model, embedding_tokenizer = load(MODEL_NAME)


def html_to_text(html):
    soup = BeautifulSoup(str(html), "html.parser")
    return soup.get_text(separator="\n", strip=True)


conn = sqlite3.connect(DB_PATH)
df = pd.read_sql(f"SELECT rowid AS source_row_id, * FROM {TABLE_NAME}", conn)
conn.close()

df["본문텍스트"] = df[TEXT_COL].map(html_to_text)
df_embed = df[df["본문텍스트"].astype(str).str.strip() != ""].reset_index(drop=True)
print("임베딩 대상 행 수:", len(df_embed))

metadata_cols = [
    "source_row_id",
    "기업명",
    "제목",
    "국가",
    "지역",
    "산업분류",
    "게시일자",
    "게시글일련번호",
    "본문내용",
    "본문텍스트",
]
metadata_cols = [col for col in metadata_cols if col in df_embed.columns]

num_batches = (len(df_embed) + BATCH_SIZE - 1) // BATCH_SIZE

for batch_id in tqdm(range(num_batches), desc="기업성공사례 임베딩"):
    emb_path = BATCH_DIR / f"embeddings_{batch_id:06d}.npy"
    meta_path = BATCH_DIR / f"metadata_{batch_id:06d}.parquet"

    if emb_path.exists() and meta_path.exists():
        continue

    start = batch_id * BATCH_SIZE
    end = min(start + BATCH_SIZE, len(df_embed))
    batch_df = df_embed.iloc[start:end].copy()

    texts = (
        "제목: " + batch_df["제목"].fillna("").astype(str) + "\n"
        "기업명: " + batch_df["기업명"].fillna("").astype(str) + "\n"
        "국가: " + batch_df["국가"].fillna("").astype(str) + "\n"
        "지역: " + batch_df["지역"].fillna("").astype(str) + "\n\n"
        "본문:\n" + batch_df["본문텍스트"].fillna("").astype(str)
    ).tolist()

    inputs = embedding_tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=MAX_LENGTH,
        return_tensors="mlx",
    )
    outputs = embedding_model(inputs["input_ids"], attention_mask=inputs["attention_mask"])

    embeddings = np.array(outputs.text_embeds).astype("float32")
    faiss.normalize_L2(embeddings)
    np.save(emb_path, embeddings)

    batch_meta = batch_df[metadata_cols].copy()
    batch_meta.insert(0, "faiss_id", range(start, end))
    batch_meta.to_parquet(meta_path, index=False)

    del inputs, outputs, embeddings, batch_df, batch_meta, texts
    gc.collect()
    mx.clear_cache()

batch_embedding_paths = sorted(BATCH_DIR.glob("embeddings_*.npy"))
batch_metadata_paths = sorted(BATCH_DIR.glob("metadata_*.parquet"))

if len(batch_embedding_paths) != len(batch_metadata_paths):
    raise ValueError("임베딩 batch와 metadata batch 개수가 다릅니다.")

index = None
for emb_path in tqdm(batch_embedding_paths, desc="FAISS 인덱스 생성"):
    embeddings = np.load(emb_path).astype("float32")
    if index is None:
        index = faiss.IndexFlatIP(embeddings.shape[1])
    index.add(embeddings)

metadata = pd.concat([pd.read_parquet(path) for path in batch_metadata_paths], ignore_index=True)
metadata["faiss_id"] = range(len(metadata))

faiss.write_index(index, str(INDEX_PATH))
metadata.to_parquet(METADATA_PATH, index=False)

config = {
    "model_name": MODEL_NAME,
    "source_db": str(DB_PATH),
    "table_name": TABLE_NAME,
    "index_path": str(INDEX_PATH),
    "metadata_path": str(METADATA_PATH),
    "batch_size": BATCH_SIZE,
    "max_length": MAX_LENGTH,
    "chunking": False,
    "html_removed": True,
    "normalized_l2": True,
    "rows": int(len(metadata)),
    "embedding_dim": int(index.d),
}
CONFIG_PATH.write_text(json.dumps(config, ensure_ascii=False, indent=2), encoding="utf-8")

print("FAISS index:", INDEX_PATH)
print("metadata:", METADATA_PATH)
print("rows:", index.ntotal)
print("dim:", index.d)
```

## 글로벌 공급망 인사이트 PDF 임베딩 파이프라인
### chunk size=1200, chunk overlap=200, batch size=16

```python
import gc
import json
import sqlite3
from pathlib import Path

import faiss
import mlx.core as mx
import numpy as np
import opendataloader_pdf
import pandas as pd
from mlx_embeddings import load
from tqdm.auto import tqdm

MODEL_NAME = "majentik/Qwen3-Embedding-0.6B-MLX-4bit"
DB_PATH = Path("data/kotra/globalSupplyInsights.db")
TABLE_NAME = "globalSupplyInsights"
PDF_DIR = Path("data/kotra/globalSupplyInsights_pdfs")
OUT_DIR = Path("data/faiss/globalSupplyInsights_pdf_qwen3_embedding")
BATCH_DIR = OUT_DIR / "batches"
INDEX_PATH = OUT_DIR / "index.faiss"
METADATA_PATH = OUT_DIR / "metadata.parquet"
CONFIG_PATH = OUT_DIR / "config.json"

BATCH_SIZE = 16
MAX_LENGTH = 32768
CHUNK_SIZE = 1200
CHUNK_OVERLAP = 200

OUT_DIR.mkdir(parents=True, exist_ok=True)
BATCH_DIR.mkdir(parents=True, exist_ok=True)

embedding_model, embedding_tokenizer = load(MODEL_NAME)


def pdf_to_markdown_path(pdf_path):
    md_path = pdf_path.with_suffix(".md")
    if md_path.exists() and md_path.stat().st_size > 0:
        return md_path

    opendataloader_pdf.convert(
        input_path=[str(pdf_path)],
        output_dir=str(pdf_path.parent),
        format="markdown",
        image_output="off",
        quiet=True,
    )
    return md_path


def extract_ntt_sn(pdf_path):
    return pdf_path.name.split("_", 1)[0]


def split_text(text, chunk_size=CHUNK_SIZE, chunk_overlap=CHUNK_OVERLAP):
    text = str(text).strip()
    if not text:
        return []

    chunks = []
    start = 0
    while start < len(text):
        end = min(start + chunk_size, len(text))
        chunk_text = text[start:end].strip()
        if chunk_text:
            chunks.append({
                "chunk_id": len(chunks),
                "chunk_start": start,
                "chunk_end": end,
                "chunk_text": chunk_text,
            })
        if end >= len(text):
            break
        start = max(end - chunk_overlap, start + 1)
    return chunks


conn = sqlite3.connect(DB_PATH)
db_df = pd.read_sql(f"SELECT rowid AS source_row_id, * FROM {TABLE_NAME}", conn)
conn.close()
db_df["게시글일련번호"] = db_df["게시글일련번호"].astype(str)

pdf_paths = sorted(PDF_DIR.glob("*.pdf"))
print("PDF 파일 수:", len(pdf_paths))

chunk_rows = []
for pdf_path in tqdm(pdf_paths, desc="PDF -> Markdown -> Chunk"):
    md_path = pdf_to_markdown_path(pdf_path)
    if not md_path.exists():
        print("markdown 변환 실패:", pdf_path)
        continue

    ntt_sn = extract_ntt_sn(pdf_path)
    matched = db_df[db_df["게시글일련번호"] == ntt_sn]
    meta = matched.iloc[0].to_dict() if len(matched) else {}
    markdown_text = md_path.read_text(encoding="utf-8", errors="ignore")

    base_row = {
        "게시글일련번호": ntt_sn,
        "PDF파일명": pdf_path.name,
        "PDF경로": str(pdf_path),
        "Markdown경로": str(md_path),
        "게시글제목": meta.get("게시글제목", ""),
        "게시물 URL": meta.get("게시물 URL", ""),
        "개요내용": meta.get("개요내용", ""),
        "공개일시": meta.get("공개일시", ""),
        "국가": meta.get("국가", ""),
        "지역": meta.get("지역", ""),
        "무역관명": meta.get("무역관명", ""),
    }

    for chunk in split_text(markdown_text):
        chunk_rows.append({**base_row, **chunk})

chunk_df = pd.DataFrame(chunk_rows)
chunk_df = chunk_df[chunk_df["chunk_text"].astype(str).str.strip() != ""].reset_index(drop=True)
print("임베딩 대상 chunk 수:", len(chunk_df))

metadata_cols = [
    "게시글일련번호",
    "게시글제목",
    "게시물 URL",
    "개요내용",
    "공개일시",
    "국가",
    "지역",
    "무역관명",
    "PDF파일명",
    "PDF경로",
    "Markdown경로",
    "chunk_id",
    "chunk_start",
    "chunk_end",
    "chunk_text",
]

num_batches = (len(chunk_df) + BATCH_SIZE - 1) // BATCH_SIZE

for batch_id in tqdm(range(num_batches), desc="글로벌 공급망 PDF chunk 임베딩"):
    emb_path = BATCH_DIR / f"embeddings_{batch_id:06d}.npy"
    meta_path = BATCH_DIR / f"metadata_{batch_id:06d}.parquet"

    if emb_path.exists() and meta_path.exists():
        continue

    start = batch_id * BATCH_SIZE
    end = min(start + BATCH_SIZE, len(chunk_df))
    batch_df = chunk_df.iloc[start:end].copy()

    texts = (
        "제목: " + batch_df["게시글제목"].fillna("").astype(str) + "\n"
        "공개일시: " + batch_df["공개일시"].fillna("").astype(str) + "\n"
        "파일명: " + batch_df["PDF파일명"].fillna("").astype(str) + "\n"
        "chunk_id: " + batch_df["chunk_id"].fillna("").astype(str) + "\n\n"
        "본문:\n" + batch_df["chunk_text"].fillna("").astype(str)
    ).tolist()

    inputs = embedding_tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=MAX_LENGTH,
        return_tensors="mlx",
    )
    outputs = embedding_model(inputs["input_ids"], attention_mask=inputs["attention_mask"])

    embeddings = np.array(outputs.text_embeds).astype("float32")
    faiss.normalize_L2(embeddings)
    np.save(emb_path, embeddings)

    batch_meta = batch_df[metadata_cols].copy()
    batch_meta.insert(0, "faiss_id", range(start, end))
    batch_meta.to_parquet(meta_path, index=False)

    del inputs, outputs, embeddings, batch_df, batch_meta, texts
    gc.collect()
    mx.clear_cache()

batch_embedding_paths = sorted(BATCH_DIR.glob("embeddings_*.npy"))
batch_metadata_paths = sorted(BATCH_DIR.glob("metadata_*.parquet"))

if len(batch_embedding_paths) != len(batch_metadata_paths):
    raise ValueError("임베딩 batch와 metadata batch 개수가 다릅니다.")

index = None
for emb_path in tqdm(batch_embedding_paths, desc="FAISS 인덱스 생성"):
    embeddings = np.load(emb_path).astype("float32")
    if index is None:
        index = faiss.IndexFlatIP(embeddings.shape[1])
    index.add(embeddings)

metadata = pd.concat([pd.read_parquet(path) for path in batch_metadata_paths], ignore_index=True)
metadata["faiss_id"] = range(len(metadata))

faiss.write_index(index, str(INDEX_PATH))
metadata.to_parquet(METADATA_PATH, index=False)

config = {
    "model_name": MODEL_NAME,
    "source_db": str(DB_PATH),
    "table_name": TABLE_NAME,
    "pdf_dir": str(PDF_DIR),
    "index_path": str(INDEX_PATH),
    "metadata_path": str(METADATA_PATH),
    "batch_size": BATCH_SIZE,
    "max_length": MAX_LENGTH,
    "chunking": True,
    "chunk_size": CHUNK_SIZE,
    "chunk_overlap": CHUNK_OVERLAP,
    "pdf_to_markdown": True,
    "normalized_l2": True,
    "rows": int(len(metadata)),
    "embedding_dim": int(index.d),
}
CONFIG_PATH.write_text(json.dumps(config, ensure_ascii=False, indent=2), encoding="utf-8")

print("FAISS index:", INDEX_PATH)
print("metadata:", METADATA_PATH)
print("rows:", index.ntotal)
print("dim:", index.d)
```

# HS 한글 품목명 기준 임베딩

`*_hs` 파일에서 `HS한글품목명목록`이 비어 있지 않은 행만 Qwen3로 임베딩합니다.

- `exportRecommendInfo_hs`
- `importRegItems_hs`
- `ovseaAuthInfo_hs`

임베딩 텍스트는 `HS한글품목명목록`만 사용하고, metadata에는 원본 컬럼을 모두 보존합니다.


```python
import gc
import json
import sqlite3
from pathlib import Path

import faiss
import mlx.core as mx
import numpy as np
import pandas as pd
from mlx_embeddings import load
from tqdm.auto import tqdm

MODEL_NAME = "majentik/Qwen3-Embedding-0.6B-MLX-4bit"
MAX_LENGTH = 32768
TEXT_COL = "HS한글품목명목록"

HS_EMBED_TARGETS = [
    {
        "stem": "exportRecommendInfo_hs",
        "table": "exportRecommendInfo_hs",
        "db_path": Path("data/kotra/exportRecommendInfo_hs.db"),
        "out_dir": Path("data/faiss/exportRecommendInfo_hs_qwen3_embedding"),
        "batch_size": 128,
    },
    {
        "stem": "importRegItems_hs",
        "table": "importRegItems_hs",
        "db_path": Path("data/kotra/importRegItems_hs.db"),
        "out_dir": Path("data/faiss/importRegItems_hs_qwen3_embedding"),
        "batch_size": 32,
    },
    {
        "stem": "ovseaAuthInfo_hs",
        "table": "ovseaAuthInfo_hs",
        "db_path": Path("data/kotra/ovseaAuthInfo_hs.db"),
        "out_dir": Path("data/faiss/ovseaAuthInfo_hs_qwen3_embedding"),
        "batch_size": 32,
    },
]

embedding_model, embedding_tokenizer = load(MODEL_NAME)


def embed_hs_target(target):
    stem = target["stem"]
    table = target["table"]
    db_path = target["db_path"]
    out_dir = target["out_dir"]
    batch_size = target["batch_size"]

    batch_dir = out_dir / "batches"
    index_path = out_dir / "index.faiss"
    metadata_path = out_dir / "metadata.parquet"
    config_path = out_dir / "config.json"

    out_dir.mkdir(parents=True, exist_ok=True)
    batch_dir.mkdir(parents=True, exist_ok=True)

    conn = sqlite3.connect(db_path)
    df = pd.read_sql(f"SELECT rowid AS source_row_id, * FROM {table}", conn)
    conn.close()

    df_embed = df[df[TEXT_COL].astype(str).str.strip() != ""].reset_index(drop=True)
    print(f"[{stem}] 전체 행 수: {len(df):,}")
    print(f"[{stem}] 임베딩 대상 행 수: {len(df_embed):,}")

    if len(df_embed) == 0:
        print(f"[{stem}] 임베딩할 행이 없어 건너뜁니다.")
        return

    metadata_cols = list(df_embed.columns)
    num_batches = (len(df_embed) + batch_size - 1) // batch_size

    for batch_id in tqdm(range(num_batches), desc=f"{stem} HS 임베딩"):
        emb_path = batch_dir / f"embeddings_{batch_id:06d}.npy"
        meta_path = batch_dir / f"metadata_{batch_id:06d}.parquet"

        if emb_path.exists() and meta_path.exists():
            continue

        start = batch_id * batch_size
        end = min(start + batch_size, len(df_embed))
        batch_df = df_embed.iloc[start:end].copy()

        texts = (
            "HS 한글 품목명:\n" + batch_df[TEXT_COL].fillna("").astype(str)
        ).tolist()

        inputs = embedding_tokenizer(
            texts,
            padding=True,
            truncation=True,
            max_length=MAX_LENGTH,
            return_tensors="mlx",
        )
        outputs = embedding_model(inputs["input_ids"], attention_mask=inputs["attention_mask"])

        embeddings = np.array(outputs.text_embeds).astype("float32")
        faiss.normalize_L2(embeddings)
        np.save(emb_path, embeddings)

        batch_meta = batch_df[metadata_cols].copy()
        batch_meta.insert(0, "faiss_id", range(start, end))
        batch_meta.to_parquet(meta_path, index=False)

        del inputs, outputs, embeddings, batch_df, batch_meta, texts
        gc.collect()
        mx.clear_cache()

    batch_embedding_paths = sorted(batch_dir.glob("embeddings_*.npy"))
    batch_metadata_paths = sorted(batch_dir.glob("metadata_*.parquet"))

    if len(batch_embedding_paths) != len(batch_metadata_paths):
        raise ValueError(f"[{stem}] 임베딩 batch와 metadata batch 개수가 다릅니다.")

    index = None
    for emb_path in tqdm(batch_embedding_paths, desc=f"{stem} FAISS 인덱스 생성"):
        embeddings = np.load(emb_path).astype("float32")
        if index is None:
            index = faiss.IndexFlatIP(embeddings.shape[1])
        index.add(embeddings)

    metadata = pd.concat(
        [pd.read_parquet(path) for path in batch_metadata_paths],
        ignore_index=True,
    )
    metadata["faiss_id"] = range(len(metadata))

    faiss.write_index(index, str(index_path))
    metadata.to_parquet(metadata_path, index=False)

    config = {
        "model_name": MODEL_NAME,
        "source_db": str(db_path),
        "table_name": table,
        "index_path": str(index_path),
        "metadata_path": str(metadata_path),
        "text_col": TEXT_COL,
        "batch_size": batch_size,
        "max_length": MAX_LENGTH,
        "chunking": False,
        "filtered_non_empty_text_col": True,
        "normalized_l2": True,
        "rows": int(len(metadata)),
        "embedding_dim": int(index.d),
    }
    config_path.write_text(json.dumps(config, ensure_ascii=False, indent=2), encoding="utf-8")

    print(f"[{stem}] FAISS index: {index_path}")
    print(f"[{stem}] metadata: {metadata_path}")
    print(f"[{stem}] rows: {index.ntotal:,}")
    print(f"[{stem}] dim: {index.d}")
    print()


for target in HS_EMBED_TARGETS:
    embed_hs_target(target)
```
